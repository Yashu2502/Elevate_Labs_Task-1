### Checking and Handling Missing Values

```python
print(data.isnull().sum())
```

- There are 24 missing values in "Income" column. To handle missing values, I am using fillna() technique and filling with median

```python
data['Income'].fillna(data['Income'].median(), inplace=True)
```

### Checking and Removing Duplicates

```python
data.duplicated().sum() 
```

### Data Formatting or Standardizing

```python
# Standarizing the text values
data['Education'] = data['Education'].str.strip().str.title()
data['Marital_Status'] = data['Marital_Status'].str.strip().str.title()
```

```python
data.head(10)
```

### Converting the dateformat into consistent datatype

```python
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], dayfirst= True)
```

```python
data.head(20)
```

###  Rename column headers to be clean and uniform (e.g., lowercase, no spaces)

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
data.columns
```

### Univariate Analysis
- We perform univariate analysis to explore distributions of individual variables.
- We will be using histplot for numerical columns and countplot for categorical columns.

```python
data.hist(bins=20, figsize = (20,15))
plt.show()
categorical_cols = ['education', 'marital_status']
for col in categorical_cols:
    sns.countplot(data=data, x=col)
    plt.xticks(rotation=45)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### Bivariate Analysis
- We use bivariate analysis to explare relationships berween variables.
- We are gonna use boxplot for cateogrical data and pairplot for numeric data

```python
# Income vs Education
sns.boxplot(data=data, x='education', y='income')
plt.xticks(rotation=45)
plt.title("Income by Education Level")
plt.show()

# Pairplot for numeric features
sns.pairplot(data[['income', 'mntwines', 'mntfruits', 'mntmeatproducts']], diag_kind='kde')
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Exporting Cleaned Dataset
